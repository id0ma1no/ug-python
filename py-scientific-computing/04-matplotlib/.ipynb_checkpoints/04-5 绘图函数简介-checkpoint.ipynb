## 绘图函数简介

### 对数坐标图

**对数轴的由来**: 将指数型增长的数据, 变为线性增长;  
**如何实现对数坐标轴**:  
* 以子图坐标系为参照, 拉伸数据坐标; 
* 将坐标轴的刻度改成对数形式;
* 由ax.transScale实现;

绘制对数坐标轴的函数:   
* semilogx(): 将X轴设置为对数轴;  
* semilogy(): 将Y轴设置为对数轴;  
* loglog(): 将两个轴都设为对数轴;  

使用对数轴之后, 函数内部自动对每个数据点取对数, 自己不需要再取对数;  
也可以自己取对数, 放置在一个线性轴上观察数据变化, 这两种做法的用途是一致的;

```python
%clf
x = np.linspace(1,110,11)
y = x
y2 = np.log(x)

fig, axes =plt.subplots(1,3, figsize=(6,2) )

# Log10Transform, 相当于对原始数据取对数
axes[0].plot(x, y, "-o", linewidth=2)
axes[1].semilogy(y, "-o", linewidth = 2)
axes[2].plot(x, y2, "-o", linewidth = 2)

# 如果y轴是对数坐标,其范围只能为正值
axes[0].set_ylim(1, 120) 
axes[1].set_ylim(1, 120) 
axes[2].set_ylim(1, 5);
```

线性轴, 对数X轴, 对数Y轴, 双对数轴

```python
%clf
w = np.linspace(0.1, 1000, 1000)
p = np.abs(1/(1+0.1j*w))
fig, axes = plt.subplots(2, 2)

functions = ("plot", "semilogx", "semilogy", "loglog")

for ax, fname in zip(axes.ravel(), functions):
    func = getattr(ax, fname)
    func(w, p, linewidth=2)
```

### 等值线图

使用等值线图查看二元函数的变化趋势;  
* 等值线: 函数值相同的各点连接成的曲线, 等高线图就是等值线的一种;  
* 等值线密集的地方表示函数值变化较大, 稀疏的地方表示函数值变化较小;

contour(): 绘制等值线图;

plt.clabel(cs) #

```python
%clf
y, x = np.ogrid[-2:2:200j, -3:3:300j] # 第0轴是Y轴,左侧的轴
z = x * np.exp( - x**2 - y**2) 

extent = [np.min(x), np.max(x), np.min(y), np.max(y)]

plt.figure(figsize=(6,3))
cs = plt.contour(z, 8, extent=extent, colors='k') # 默认已经将数组镜像,所以不需要设置origin='lower'

plt.clabel(cs,fontsize=8,fmt='%1.1f',colors='r') # 为轮廓集对象添加标签
plt.contourf(x.reshape(-1), y.reshape(-1), z, 8) #
plt.colorbar()
plt.show()
```

如果一个一元隐函数无法进行离散点选取, 可以使用contour()绘制它的函数曲线;  
* 因为任意一个一元隐函数(如果存在的话)都是一个二元方程的解: $f(x,y)={(x^2+y^2)}^4 -{(x^2-y^2)}^2$;  
* 使用等值线绘制隐函数曲线, 从而获取离散点, 使用plot重新绘图;

```python
y, x = np.ogrid[-1.5:1.5:200j, -1.5:1.5:200j]
f = (x**2 + y**2)**4 - (x**2 - y**2)**2

plt.figure(figsize=(4.5, 2))


plt.subplot(121)
extent = [np.min(x), np.max(x), np.min(y), np.max(y)]

cs = plt.contour(f, extent=extent, levels=[0, 0.1],
     colors=["b", "r"], linestyles=["solid", "dashed"], linewidths=[2, 2])
plt.subplot(122)
for c in cs.collections: # Path(array(),None)
    data = c.get_paths()[0].vertices # 获取离散数据array()
    plt.plot(data[:,0], data[:,1], color=c.get_color()[0],  linewidth=c.get_linewidth()[0])

# plt.clabel(cs) #先获取数据再设置标签
```

### 四边形网格

pcolormesh(X,Y,C): 绘制四边形网格以及为其填充颜色;  
* X: X中上下左右相邻的四个元素共同决定一个四边形的X坐标;
* Y: Y中上下左右相邻的四个元素共同决定一个四边形的Y坐标;
* C: C决定每个四边形的颜色, 其形状比X和Y少一行一列;  
* X与Y的形状必须相同; 如果Z与X, Y的形状相同, 将剔除其最后一行和最后一列?

```python
%clf

X = np.array([[0, 1, 2], 
              [0, 1, 2]])
Y = np.array([[0, 0.2, 0],
              [1, 0.8, 1]])
Z = np.array([[0.5, 0.8]])

# 与上面输出相同: 如果Z的形状与X,Y相同, 则剔除Z的最后一行和最后一列
# Z = np.array([[0.5, 0.8, 0.2],[0.8, 0.5, 0.2]]) 

plt.figure(figsize=(4,2))
plt.plot(X.ravel(), Y.ravel(), "ko")
plt.pcolormesh(X, Y, Z)
plt.margins(0.1);
plt.axis('equal')
```

使用pcolormesh()绘制复数平面上的坐标变换;  
* 实部做X轴, 虚部做Y轴, 复数的模表示颜色, 函数自动去除C模矩阵的最后一行和最后一列; 

```python
%clf
def make_mesh(n):
    x, y = np.mgrid[-10:0:n*1j, -5:5:n*1j]
    s = x + 1j*y #
    z = (2 + s) / (2 - s)
    return s, z

fig, axes = plt.subplots(2, 2, figsize=(6, 6))
axes = axes.ravel()
for ax in axes:
    ax.set_aspect("equal")
    
s1, z1 = make_mesh(5)
s2, z2 = make_mesh(20)
axes[0].pcolormesh(s1.real, s1.imag, np.abs(s1)) #计算复数的模, sqrt(x^2+y^2)
axes[1].pcolormesh(z1.real, z1.imag, np.abs(s1))
axes[2].pcolormesh(s2.real, s2.imag, np.abs(s2)) # 如果四边形个数过大, 设置rasterized=True
axes[3].pcolormesh(z2.real, z2.imag, np.abs(s2));
```

`diffusion.txt`文件的结构:  
* points: (N_points,2), N个点的坐标, 坐标点的个数并不是三角形的个数;
* truangles: (N_triangles,3), 三角形三个顶点在points数组中的下标;
* values: (N_triangles,1), 每个三角形对应的温度;

```python
kk 4
```

```python
with open("data/diffusion.txt") as f:
    data = {"points":[], "triangles":[], "values":[]}
    values = None
    for line in f:
        line = line.strip()
        if not line: #如果不是空行, 继续
            continue 
        if line.startswith("#"):
            values = data[line[1:]]
            continue
        values.append([float(s) for s in line.split()])
        
data = {key:np.array(data[key]) for key in data}
```

使用tripcolor()和tricontour()绘制三角网格和等值线

```python
X, Y = data["points"].T
triangles = data["triangles"].astype(int)
values = data["values"].squeeze()

fig, ax = plt.subplots(figsize=(12, 4.5))
ax.set_aspect("equal")

mapper = ax.tripcolor(X, Y, triangles, values, cmap="gray")
plt.colorbar(mapper, label="Temperature")

plt.triplot(X, Y, triangles, lw=0.5, alpha=0.3, color="k") 

Xc = X[triangles].mean(axis=1)
Yc = Y[triangles].mean(axis=1)
plt.tricontour(Xc, Yc, values, 10) 
```

### 箭头图

quiver(): 绘制用大量箭头表示的矢量场;  
绘制二元函数的梯度场: $$f(x,y)=x e^{{x^2}-{y^2}}$$

```python
%clf
def f(x, y):
    return x * np.exp(- x**2 - y**2)
# 计算函数在x,y处的偏导数
def vec_field(f, x, y, dx=1e-6, dy=1e-6):
    x2 = x + dx
    y2 = y + dy
    v = f(x, y)
    vx = (f(x2, y) - v) / dx #计算对x的偏导数,y是常数
    vy = (f(x, y2) - v) / dy 
    return vx, vy

X, Y = np.mgrid[-2:2:20j, -2:2:20j]
C = f(X, Y)
U, V = vec_field(f, X, Y)
plt.quiver(X, Y, U, V, C)
plt.colorbar();
plt.gca().set_aspect("equal")
```

```python
%clf
def f(x,y):
    return (y-x)/np.sqrt(x**2+y**2)
# 函数的偏导数就代表箭头矢量
def vec_field(f, x, y, dx=1e-6, dy=1e-6):
    x2 = x + dx
    y2 = y + dy
    v = f(x, y)
    vx = (f(x2, y) - v) / dx
    vy = (f(x, y2) - v) / dy 
    return vx, vy

X, Y = np.mgrid[1:3:6j, 1:3:6j]
C = f(X, Y) #函数值的大小用颜色进行区分
U, V = vec_field(f, X, Y)
plt.quiver(X, Y, U, V, C)
plt.colorbar();
plt.gca().set_aspect("equal")
```

使用箭头表示参数曲线的切线方向

```python
%clf
n = 40
arrow_size = 16
t = np.linspace(0, 1, 1000)
x = np.sin(3*2*np.pi*t)
y = np.cos(5*2*np.pi*t)
line, = plt.plot(x, y, lw=1)

lengths = np.cumsum(np.hypot(np.diff(x), np.diff(y)))
length = lengths[-1]
arrow_locations = np.linspace(0, length, n, endpoint=False)
index = np.searchsorted(lengths, arrow_locations)
dx = x[index + 1] - x[index]
dy = y[index + 1] - y[index]
ds = np.hypot(dx, dy)
dx /= ds
dy /= ds
plt.quiver(x[index], y[index], dx, dy, t[index],
          units="dots", scale_units="dots", 
          angles="xy", scale=1.0/arrow_size, pivot="middle",
          edgecolors="black", linewidths=1,
          width=1, headwidth=arrow_size*0.5, 
          headlength=arrow_size, headaxislength=arrow_size, 
          zorder=100)
plt.colorbar()
plt.xlim([-1.5, 1.5])
plt.ylim([-1.5, 1.5]);
```

使用quiver()绘制神经网络结构示意图

```python
%clf
levels = [4, 5, 3, 2]
x = np.linspace(0, 1, len(levels))

for i in range(len(levels) - 1):
    j = i + 1
    n1, n2 = levels[i], levels[j]
    y1, y2 = np.mgrid[0:1:n1*1j, 0:1:n2*1j]
    x1 = np.full_like(y1, x[i])
    x2 = np.full_like(y2, x[j])
    plt.quiver(x1, y1, x2-x1, y2-y1, 
              angles="xy", units="dots", scale_units="xy", 
              scale=1, width=2, headlength=10,
              headaxislength=10, headwidth=4)
    
yp = np.concatenate([np.linspace(0, 1, n) for n in levels])
xp = np.repeat(x, levels)
plt.plot(xp, yp, "o", ms=12)
plt.gca().axis("off")
plt.margins(0.1, 0.1);
```
