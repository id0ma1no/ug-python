## Artist对象

**artist对象**: 用于处理图元的绘制和布局;  

* 简单类型(基本图元): Line2D, Rectangle, Text, AxesImage
* 容器类型(复合图元): Axis, Axes, Figure

绘图流程:  
1. 创建图形对象(Figure)
1. 在图形对象之中创建子图对象(轴对象, Axes)
1. 调用子图对象的各种方法创建基本图元

使用函数plt.figure()创建图形对象;  
使用fig.add_axes()为图形对象添加子图对象;  
* 可以使用四元组指定子图的位置: `[left, bottom, width, height]`
* `projection` : 指定轴的投影类型;

```python
%clf
fig = plt.figure(figsize=(2,2))
ax = fig.add_axes([0, 0, 0.4, 0.4])
ax1 = fig.add_axes([0.5, 0.5, 0.5, 0.5], polar=True)
plt.show()
```

ax.plot(): 绘制二维曲线, 返回一个Line2D对象列表

```python
pi = np.pi
ax.plot([1, 2, 3], [1, 2, 1])
ax1.plot([0, pi/2, pi, 3*pi/2, 0], [1, 1, 1,1,1])
plt.show()
```

通过 set_xlabel() 设置X轴上的标题

```python
ax.set_xlabel("直角坐标系", fontdict={'fontname': 'simhei', 'fontsize':8})
ax1.set_xlabel("极坐标系", fontdict={'fontname': 'simhei', 'fontsize':8})
plt.show()
```

查看set_xlabel()的源码: ` return self.xaxis.set_label_text(xlabel, fontdict)`;  
`ax`的`xaxis`是一个**XAxis**对象, 其`label属性`是一个**Text**对象, Text对象的`_text属性`是我们需要设置的;

```python
%P ax.xaxis; ax.xaxis.label; ax.xaxis.label._text; ax.get_xlabel()
```

### Artist对象的属性

Figure对象和Axes对象的都有`patch属性`, 它是一个Rectangle对象, 通过该属性可以设置图形对象的背景色和透明度

```python
# plt.%clf
fig.patch.set_color("gray"); fig.patch.set_alpha(0.5)
ax.patch.set_color('green'); ax.patch.set_alpha(.3)
ax1.patch.set_color('red'); ax.patch.set_alpha(.3); 
# fig.canvas.draw() #跟新界面显示
plt.show()
```

Artist对象的所有属性都可以通过相应的`ax.get_*()`和`ax.set_*()`进行读写;  
`plt.getp()`: 一次性获取图元的属性;  
`ax.set()`或`plt.setp()`: 一次性设置图元的属性;

```python
line = ax.lines[0] # 必须取一条线条, 不能去掉[0]
line1= ax1.lines[0]
```

```python
# plt.getp(line)
line.set(ls='None', marker='o', ms=15)
line1.set(ls='None',marker='s', ms=15)
plt.show()
```

### Figure容器

在构成图表的各种Artist对象中, 最底层的Artist对象是Figure, 它包含图表中的所有元素;  

Figure对象之上便是Axes对:

* fig.add_axes(): 往图表中添加子图, 并返回一个Axes对象;
* fig.add_subplot(): 往图表中添加子图, 并返回一个AxesSubplot对象, AxesSubplot类是Axes类的派生类;
* 虽然可以使用fig.axes属性进行子图的添加操作, 但最好不要这样, 而是使用fig.add_axes()方法

```python
%clf
fig = plt.figure( figsize = (4,2) )
ax1 = fig.add_subplot(211)
ax2 = fig.add_axes([0.12,0.12,0.78,0.3])

# 不要使用fig.axes返回的列表直接操作子图
%P ax1 in fig.axes; ax2 in fig.axes 

# 子图对象的遍历
for ax in fig.axes:
    ax.grid(True)
plt.show()
```

图表对象是一个容器类型的Artist对象, 可以拥有自己的文本, 线条, 图像;  
* `transform`: 设置图元所使用的坐标系;

```python
%clf
from matplotlib.lines import Line2D
fig = plt.figure( figsize = (2,1) )
line1 = Line2D([0, 1], [0, 1], transform=fig.transFigure, figure=fig, color="r")
line2 = Line2D([0, 1], [1, 0], transform=fig.transFigure, figure=fig, color="g")
fig.lines.extend([line1, line2])
```

一般情况下, 不要直接使用lines或patches属性直接对Artist对象进行操作;  
而应该使用add_line()或add_patch()等方法对Artist对象进行操作;

```python
%clf
# 将矩形对象rect添加进子图对象ax 
fig = plt.figure(figsize = (1,1))
ax = fig.add_axes([0,0,1,1])
rect = plt.Rectangle((1,1), width=1, height=1, color='red', alpha=0.2) #矩形对象
ax.add_patch(rect) # rect.get_axes() is ax -->True

# 1) x轴的范围 2) rect的范围, 在x轴的范围内无法显示 3) 自动调整轴的范围以使其显示rect 4)调整轴范围之后的想轴范围
%P ax.get_xlim(); ax.dataLim.bounds; ax.autoscale_view(); ax.get_xlim() # 于是X轴可以完整显示rect

# 取消rect与坐标轴之间的空白间隙
ax.set_xlim(1,2); ax.set_ylim(1,2);
```

轴刻度设置的综合示例:  
* 配置X轴的刻度线的位置和文本:  
    * 主刻度为pi/4, 副刻度为pi/20;
    * 主刻度上的文本用$\pi$显示
* 开启副刻度线;

FuncFormatter( pi_formatter ): 使用自定义函数计算文本格式;  
* func()应该接收两个参数(刻度值, 位置), 然后返回一个刻度标签;

```python
from fractions import Fraction
from matplotlib.ticker import MultipleLocator, FuncFormatter 
def pi_formatter(x, pos): 
    frac = Fraction(int(np.round(x / (np.pi/4))), 4)
    d, n = frac.denominator, frac.numerator
    # 如果分式为0的话
    if frac == 0:
        return "0"
    #如果分式为1的话
    elif frac == 1:
        return "$\pi$"
    #如果分母为1的话, 直接返回分子
    elif d == 1: 
        return r"${%d} \pi$" % n
    #如果分子为1的话,返回分母的倒数
    elif n == 1: 
        return r"$\frac{\pi}{%d}$" % d 
    return r"$\frac{%d \pi}{%d}$" % (n, d) 
    #备注: Latex字符串中可以使用格式占位符
```

```python
%clf
x = np.arange(0, 4*np.pi, 0.01)
fig, ax = plt.subplots(figsize=(6,3))
plt.plot(x, np.sin(x), x, np.cos(x))
# 设置两个坐标轴的范围
plt.ylim(-1.5,1.5); plt.xlim(0, np.max(x))
#开启网格线
plt.grid() 

# 设置图的底边距
plt.subplots_adjust(bottom = 0.15)

# 主刻度为pi/4
ax.xaxis.set_major_locator( MultipleLocator(np.pi/4) ) #刻度的数值间隔
# 主刻度文本用pi_formatter函数计算
ax.xaxis.set_major_formatter( FuncFormatter(pi_formatter) )

# 副刻度为pi/20
ax.xaxis.set_minor_locator( MultipleLocator(np.pi/20) ) 

# 设置主刻度文本的大小
for tick in ax.xaxis.get_major_ticks():
    tick.label1.set_fontsize(10)
```
