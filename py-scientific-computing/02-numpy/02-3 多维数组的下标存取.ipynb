#### 三维数组在Numpy是如何表示的

三维数组应该是三维空间中的一些离散点, 如果想要在二维平面上表示出来, 只能使用投影的方式;  
在Numpy中, 所有的三维数组都被投影在了x轴上(yoz面), 

下标对象是一个元组, 元组的长度等于多维数组的维数;   
从投影的角度理解**三维数组(3,4,5)**的含义:  
* Numpy中的空间坐标系: X轴垂直于纸面于O点(X轴垂直于纸面向上), 建立三维直角坐标系, 将其投影到纸面, 向下延伸的方向是Y轴, 向右延伸的方向是Z轴,  Y轴与Z轴之间的正向夹角是90度;
* 沿X轴向原点看去, 会遇到3个二维数组(在三个平面上), 每个二维数组的形状都是(4,5), 将这3个数组投影到yoz面上, 并依次向下错开, 就形成了三维数组的平面表示; 

沿x轴的方向向原点看过去, 所遇到的第一个数组位于平面表示的最下面, 最后一个数组位于平面表示的最上面;  
将第二个数组往上平移到第一个数组的位置, 重叠的数字其y和z坐标相同, 而x坐标不同; 例如0,20,40都是分别位于x轴上的三个元素, 其yz坐标相同, 而x位于不同的纸面高度;

#### 下标元组中只有整数数组

```python
i0 = np.array([[1, 2, 1], [0, 1, 0]])
i1 = np.array([[[0]], [[1]]])
i2 = np.array([[[2, 3, 2]]])
b = a[i0, i1, i2]
%C 5 a.shape; i0.shape; i1.shape; i2.shape; b.shape
```

shape属性不足的部分在前面加1补齐

```python
for ary in [i0,i1,i2]:
    if ary.ndim != a.ndim:
        ary.shape = (1,)*(a.ndim-ary.ndim)+ary.shape
    print(ary.shape)
```

**数组能够广播的条件**: 上面的每一列元素挖去1之外, 所有的元素都必须相同, 如果不同则不能广播;  
经过广播之后, 3个数组的形状相同, 并且这3个数组相同位置处的元素, 共同决定了该处结果数组的元素;

```python
ind0, ind1, ind2 = np.broadcast_arrays(i0, i1, i2)
%C 5 ind0; ind1; ind2; b #22是a中第一个数组的0行2列
```

#### 下标元组中既有整数数组又有切片

下标含有切片时, 应遵循一般的索引规则: 索引在前和索引在后的形式(**后面数组替换前面数组的空框形式**);  
1. 切片在后:  
    * 将整数数组进行广播, 写出广播数组的结构表示为`空框形式`;  
    * 将每个空框换成切片代表的一维数组;
1. 切片在前:  
    * 将切片表示为`空框形式`; 
    * 令x分别为切片的每个索引, 配合其余数组得到的部分结果数组;  
    * 将部分结果数组放入切片的空框形式;

切片位于整数数组们的两侧: 原位组合切片的形状和广播数组的形状, 得结果数组的形状

```python
c = a[:, i0, i1]
ind0,ind2 = np.broadcast_arrays(i0, i1)  #数组a的第一个轴的长度是3, 切片的长度是3
%C 5 ind0.shape; ind2.shape; c.shape; c.shape == (3,)+ind0.shape
```

```python
%C 5 ind0; ind1; a; c
```

整数数组不连续, 切片在中间时结果数组的形状: 将切片的形状放在其他数组广播之后的形状的后面;

```python
d = a[i0, :, i1] #数组a第二个轴的长度是4, 切片的长度也是4
%C ind0.shape; d.shape
```

```python
%C ind0; ind1; a; d
```

### 一个复杂的例子

已知整数`L`, 数组`v`及其形状: 
* `v.shape = I,J,K`  
* `idx.shape = I,J`, idx的每个值都是`0到K-L`之间的整数  

求满足`r[i,j,:] = v[i,j,idx[i,j]:idx[i,j]+L]`的数组`r`

```python
I, J, K, L = 3, 4, 5, 3

# 数组v: 在z轴上, 每层的值都是该层的高度
_, _, v = np.mgrid[:I, :J, :K]

# 随机产生数组idx, 它的每个元素都在0到K-L之间
idx = np.random.randint(0, K - L, size=(I, J))

# 创建第0轴和第1轴的下标数组
idx_i, idx_j, _ = np.ogrid[:I, :J, :K]

# 创建第2轴的下标数组
idx_k = idx[:, :, None] + np.arange(L)

r = v[idx_i, idx_j, idx_k]    

%C 3 v; idx; idx_i; idx_j; idx_k; r ;;\
                                    v.shape; idx.shape; idx_i.shape; idx_j.shape; idx_k.shape; r.shape
```
