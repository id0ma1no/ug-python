rolling_*()函数, 对序列中相邻的N个元素进行移动窗口运算

```python
#%fig=中值滤波和移动平均
t = np.linspace(0, 10, 400)
x = np.sin(0.5*2*np.pi*t)
x[np.random.randint(0, len(t), 40)] += np.random.normal(0, 0.3, 40)
s = pd.Series(x, index=t)
s_mean = pd.rolling_mean(s, 5, center=True)
s_median = pd.rolling_median(s, 5, center=True)
#%hide
ax = s.plot(label=u"噪声信号")
(s_median - 0.2).plot(ax=ax, label=u"中值滤波")
(s_mean + 0.2).plot(ax=ax, label=u"移动平均")
ax.legend(loc="best", ncol=3, bbox_to_anchor=(0., 1.02, 1., .102));
```

expanding_*() 计算历史最大值、平均值、最小值

```python
np.random.seed(42)
x = np.cumsum(np.random.randn(400))
x_max = pd.expanding_max(x)
x_min = pd.expanding_min(x)
x_mean = pd.expanding_mean(x)
#%hide
df = pd.DataFrame(np.c_[x, x_max, x_mean, x_min], 
                  columns=["data", "expanding max", "expanding mean", "expanding min"])
ax = df.plot()
ax.legend(ncol=2, loc="upper left");
```

>**7位字符集(表示128个字符):**
>>**ASCII码:** 美国标准信息交换码, American Standard Code for Information Interchange, 为英语通信所设计, 它由128个字符(7个二进制位)组成，包括大小写字母、数字0-9、标点符号、非打印字符（换行符、制表符等4个）以及控制字符（退格、响铃等）组成。

>**8位字符集(表示256个字符):** 
>>**IBM字符集:** 它把值为128-255之间的字符用于画图和画线，以及一些特殊的欧洲字符。 
>>**ISOLatin-1:** 它把位于128-255之间的字符用于拉丁字母表中特殊语言字符的编码

>**16位字符集(表示2^16个字符):** 
>>**Unicode字符集:** 用2个字节可以编码地球上几乎所有地区的文字。

>**通用转换格式:**
>>**utf:** Unicode Transformation Format, 因为UNICODE比ASCII占用大一倍的空间, 所以对可以用ASCII表示的字符使用UNICODE并不高效,所以就出现了一些中间格式的字符集, 如: UTF-7, UTF-7.5, UTF-8,UTF-16, 以及 UTF-32

Python中包含两种字符串, 字节字符串, Unicode字符串,

**字节字符串** 就是一个包含字节的列表, 当需要的时候，Python根据电脑默认的locale设置将字节转化成字符。

创建一个字节字符串:
  
    byteString = "hello world!"


创建一个Unicode字符串:
        
    unicodeString = u"hello Unicode world!"


将一个字节字符串转成Unicode字符串然后再转回来:
    
    s = "hello normal string"
    u = s.decode("UTF-8" )
    backToBytes = u.encode( "UTF-8" )

用字符串表示分类信息时, 会有大量相同的字符串, 将其转换为**分类序列**可以节省内存, 如下Contour, Depth, Gp由字符串类型变为**分类类型**

```python
df_soil = pd.read_csv("Soils.csv", usecols=[2, 3, 4, 6])
print df_soil.head(), "\n"
print df_soil.dtypes
```

```python
for col in ["Contour", "Depth", "Gp"]:
    df_soil[col] = df_soil[col].astype("category")

print df_soil.head(), "\n"
print df_soil.dtypes 
```

名称空间cat, 保存了与分类序列相关的各种属性和方法

cat.categories是保存所有分类的Index对象

```python
Gp = df_soil.Gp
print Gp.cat.categories
```

cat.codes是将类型用整数表示

分类数据分为有序和无序两种, 可以用as_ordered(), as_unordered()相互转化

```python
depth = df_soil.Depth
depth.cat.as_ordered().head()
#  最后一行使用<连接表示有序分类
```
