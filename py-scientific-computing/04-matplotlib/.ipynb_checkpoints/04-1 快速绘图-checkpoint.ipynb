# matplotlib-绘制精美的图表

## 快速绘图

### 使用pyplot模块绘图

```python
#%fig=使用pyplot模块快速将数据绘制成曲线图
import matplotlib.pyplot as plt 

x = np.linspace(0, 10, 1001)
y = np.sin(x)
z = np.cos(x)

plt.figure(figsize=(4,2)) 

#  plt.plot 实际上是在子图对象上绘图
plt.plot(x,y,label="$sin(x)$",color="red",linewidth=2)
#  关键字参数可大写可小写, 也可以缩写 linewidth=lw
plt.plot(x,z,"b--",label="$cos(x^2)$")


plt.title("PyPlot First Example") 

plt.xlabel("Time(s)") 
plt.ylabel("Volt") 
plt.ylim(-1.2,1.2) 
plt.legend() 

plt.show() 
```

pyplot 模块下的 savefig()函数可以保存图像

```python
plt.savefig("test.png", dpi=80) 
# dpi 表示每英寸上的像素数 长320,宽160
```

> 如果关闭了图表窗口，则无法使用`savefig()`保存图像。实际上不需要调用`show()`显示图表，可以直接用`savefig()`将图表保存成图像文件。使用这种方法可以很容易编写批量输出图表的程序。    

```python
import io
buf = io.BytesIO() # 创建一个用来保存图像内容的BytesIO对象
plt.savefig(buf, fmt="png") # 将图像以png格式保存进buf中
buf.getvalue()[:20] # 显示图像内容的前20个字节
```

### 配置属性

每个绘图元素(线条,文字,刻度等)都是一个对象, 可以通过set_*()方法设置这些对象的属性值, 或者直接使用setp()函数一次设置多个属性

```python
#%fig[1x3]=配置绘图对象的属性
plt.figure(figsize=(3, 1.5))
x = np.arange(0, 5, 0.1)

# plot返回一个元素类型为Line2D的列表
line = plt.plot(x, 0.05*x*x)[0] 

# 调用Line2D对象的set_*()方法设置属性值
line.set_alpha(0.5) 
```

```python
# lines 是有两个Line2D对象的列表
lines = plt.plot(x, np.sin(x), x, np.cos(x))
```

```python
# 使用 setp() 函数可以一次设置某个对象的不同属性
plt.setp(lines, color="r", linewidth=4.0);
```

绘图对象的get_*()方法和getp()函数 得到对象(只能一个)的属性值

* 指定属性名: 返回对象关于此属性的值
* 不指定属性名: 返回该对象的所有属性和值

```python
print line.get_linewidth()
print plt.getp(lines[0], "color") 
# lines[0],不能是lines
# 指定属性名color, 返回color属性的值
```

```python
f = plt.gcf()
%omit plt.getp(f) #getp函数只有对象参数, 没有属性参数
```

```python
print plt.getp(f, "axes") # 带子图属性参数 
print plt.getp(f, "axes")[0]
print plt.gca()
```

```python
lines_all = plt.getp(plt.gca(), "lines")
print lines_all, "\n", lines_all[0], "\n",line
```

### 绘制多子图

**subplot(numRows, numCols, plotNum)**

plotNum 指定创建子图的区域, 图表区域按从从左到右,从上到下的顺序编号

```python
#%fig[1x2]=在Figure对象中创建多个子图
plt.figure(figsize=(5,2.5)) 
for idx, color in enumerate("rgbyck"):  
    plt.subplot(321+idx, axisbg=color) # axisbg 设置子图背景色
```

```python
plt.figure(figsize=(3,2))
plt.subplot(221) # 第一行的左图
plt.subplot(222) # 第一行的右图
plt.subplot(212) # 第二整行;
```

```python
#关闭所有的图形对象
plt.close("all")
```

```python
# 同时在多幅图表、多个子图中进行绘图
plt.figure(1, figsize=(4,2)) # 创建图表1
plt.figure(2, figsize=(4,2)) # 创建图表2
ax1 = plt.subplot(121) # 在图表2中创建子图1
ax2 = plt.subplot(122) # 在图表2中创建子图2

x = np.linspace(0, 3, 100)
for i in xrange(5):
    plt.figure(1)  # 选择图表1
    plt.plot(x, np.exp(i*x/3))
    plt.sca(ax1)   # 选择子图1, 同时选择了图表2 
    plt.plot(x, np.sin(i*x))
    plt.sca(ax2)  # 选择图表2的子图2
    plt.plot(x, np.cos(i*x))
```

> 也可以不调用`sca()`指定当前子图，而直接调用`ax1`和`ax2`的`plot()`方法绘图。

subplot可以一次生成多个子图, 并返回表示图表对象和子图对象的数组

```python
#%nofig
fig, axes = plt.subplots(2, 3)
[a, b, c], [d, e, f] = axes
print axes.shape
print b
```

使用subplot2grid()创建表格布局

subplot2grid(shape, loc,rowspan=1, colspan=1, **kwargs) 
* shape 表示将图标分割成的网格数
* rowspan 表示子图所占据的行数

```python
fig = plt.figure(figsize=(3, 3))
ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=2)
ax2 = plt.subplot2grid((3, 3), (0, 2), rowspan=2)
ax3 = plt.subplot2grid((3, 3), (1, 0), rowspan=2)
ax4 = plt.subplot2grid((3, 3), (2, 1), colspan=2)
ax5 = plt.subplot2grid((3, 3), (1, 1));

for idx, ax in enumerate(fig.axes, 1): # 1可以表示枚举的起点
    ax.text(0.5, 0.5, "ax{}".format(idx), ha="center", va="center", fontsize=8)
```

显示系统中所有的中文字体名

```python
import os
from os import path

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)
plt.subplots_adjust(0, 0, 1, 1, 0, 0)
plt.xticks([])
plt.yticks([])

x, y = 0.05, 0.05
fonts = [font.name for font in fontManager.ttflist if 
             path.exists(font.fname) and os.stat(font.fname).st_size>1e6] #❶
font = set(fonts)
dy = (1.0 - y) / (len(fonts) // 4 + (len(fonts)%4 != 0))

for font in fonts:
    t = ax.text(x, y + dy / 2, u"中文字体", 
                {'fontname':font, 'fontsize':14}, transform=ax.transAxes) #❷
    ax.text(x, y, font, {'fontsize':12}, transform=ax.transAxes)
    x += 0.25
    if x >= 1.0:
        y += dy
        x = 0.05
plt.show()
```

```python
from matplotlib.font_manager import FontProperties
fig = plt.figure(figsize=(2, 1))

font = FontProperties(fname=r"c:\windows\fonts\simsun.ttc", size=14) #❶
t = np.linspace(0, 10, 1000)
y = np.sin(t)
plt.close("all")
plt.plot(t, y)
plt.xlabel(u"时间", fontproperties=font) #❷
plt.ylabel(u"振幅", fontproperties=font)
plt.title(u"正弦波", fontproperties=font)
plt.show()
```
