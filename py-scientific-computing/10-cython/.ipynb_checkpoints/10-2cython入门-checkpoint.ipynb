```python
def two_point_distance(point1=(0,0,0,0), point2=(1,2,2,4)):
    """任意维度两点间的距离."""
    d = 0.0
    for component1,component2 in zip(point1,point2):
        d = d + (component2-component1)**2
    return round(d**0.5,3)

def distance_matrix(X):
    """二维数组的距离矩阵."""
    m, n = X.shape
    D = np.empty((m, m), dtype=np.float)
    for i in range(m):
        for j in range(i, m):
            D[j,i] = D[i,j] = two_point_distance(X[i],X[j])
    return np.round(D,3)
```

```python
def pairwise_dist_cython(X):
    m, n = X.shape
    D = np.empty((m, m), dtype=np.float)
    for i in range(m):
        for j in range(i, m):
            d = 0.0
            for k in range(n):
                tmp = X[i, k] - X[j, k]
                d += tmp * tmp
            D[i, j] = D[j, i] = d ** 0.5
    return D
```

```python
np.random.seed(42)
X = np.random.rand(200,3)
```

```python
from scipy.spatial.distance import pdist, squareform
D_sp = squareform(pdist(X)) #153 µs ± 1.4
D_book = pairwise_dist_python(X) #193 ms
D_br = distance_matrix(X) #194 ms
```

```python
%timeit pairwise_dist_cython(X)
np.allclose(pairwise_dist_cython(X), pairwise_dist_python(X))
```

```python
%%cython
import numpy as np
import cython
from libc.math cimport sqrt

@cython.boundscheck(False)
@cython.wraparound(False)
def pairwise_dist_cython2(double[:, ::1] X):
    cdef int m, n, i, j, k
    cdef double tmp, d
    m, n = X.shape[0], X.shape[1]    
    cdef double[:, ::1] D = np.empty((m, m), dtype=np.float64)
    for i in range(m):
        for j in range(i, m):
            d = 0.0
            for k in range(n):
                tmp = X[i, k] - X[j, k]
                d += tmp * tmp
            D[i, j] = D[j, i] = sqrt(d)
    return np.asarray(D)
```

### 使用cdef关键字声明变量类型

> **WARNING**

> 请注意这里使用`cdef`定义的三个全局变量为C语言的全局变量，并不能在Python中通过编译之后的扩展模块获取它们的值。

```python
%%cython -a
cdef double a = 1.0
cdef double b = 2.0
cdef double c = a + b
```

```python
%%cython -a

cdef double s = 0
a = 3.0
s = s + a #❶
s = s + <double>a #❷
```

```python
%%cython -a
cdef list clist = [1000, 2, 3]
cdef int cindex = 0
clist[cindex] #❶

pylist = [1000, 2, 3]
pyindex = 0
pylist[pyindex] #❷
```

### 使用def定义函数

```python
%%cython -a

def py_square_add(double x, double y):
    return x*x + y*y
```

```python
%%cython -a

def sum_list(list alist): 
    cdef double s = 0
    cdef int i = 0
    for i in range(len(alist)):
        s += <double>alist[i]
    return s
```

### 使用cdef定义C语言函数

> **TIP**

> 如果不声明`cdef`函数的返回值类型，则其类型为Python对象。

```python
%%cython -a
cdef double c_square_add(double x, double y):
    return x*x + y*y

cdef double a = c_square_add(1.0, 2.0)
```

```python
%%cython -a
cpdef double cp_square_add(double x, double y):
    return x*x + y*y

cp_square_add(1.0, 2.0)
```
